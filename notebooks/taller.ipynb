```python
import ipywidgets as widgets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
```

```python
archivo1 = pd.read_csv('data/xaa.csv', encoding = 'latin-1')
archivo2 = pd.read_csv('data/xab.csv',encoding = 'latin-1')
archivo3 = pd.read_csv('data/xac.csv',encoding = 'latin-1')
archivo4 = pd.read_csv('data/xad.csv',encoding = 'latin-1')
archivo5 = pd.read_csv('data/xae.csv',encoding = 'latin-1')
archivo6 = pd.read_csv('data/xaf.csv',encoding = 'latin-1')
archivo7 = pd.read_csv('data/xag.csv',encoding = 'latin-1')
data = pd.concat([archivo1, archivo2, archivo3, archivo4, archivo5, archivo6, archivo7])
```

```python
def convertidor(x):
    if (x == "Sin información" or x == " "):
        x = 0
    x = int(x)
    return x
```

```python
data['peso'] = data['peso'].apply(convertidor).astype('float')
```

```python
from IPython.display import display
w = widgets.Dropdown(
    options=[u'Salinas', u'Santa Elena', u'Guayaquil', u'La Libertad',u'Mejia', u'Pangua', u'Sigchos'],
    value = u'Salinas',
    description='Canton:',
    disabled=False,
)
display(w)
```

```python
data[data['cant_insc']== w.value]
```

```python
region_costa = ['Santa Elena ','Esmeraldas','Guayas','Santo Domingo de los Tsáchilas',  'Los Ríos','Manabí','El Oro']
region_sierra = ['Tungurahua', 'Pichincha', 'Loja', 'Imbabura', 'Chimborazo', 'Cotopaxi', 'Carchi', 'Cañar', 'Bolívar', 'Azuay']

```

```python
data['dia_insc'] = data['dia_insc'].apply(convertidor).astype('int')
```

```python
w = widgets.Dropdown(
    options= region_costa,
    description='Provincia:',
    disabled=False,
)
display(w)
```

```python
data[(data['prov_insc']== w.value) & (data['dia_insc'] > 15)]
```

```python
w = widgets.Dropdown(
    options= region_sierra,
    description='Provincia:',
    disabled=False,
)
display(w)

z = widgets.Dropdown(
    options= ['Cesárea', 'Normal'],
    value = 'Normal',
    description='Tipo de Parto',
    disabled=False,
)
display(z)
```

```python
data[(data['prov_insc'] == w.value) & (data['tipo_part'] == z.value)]
```
